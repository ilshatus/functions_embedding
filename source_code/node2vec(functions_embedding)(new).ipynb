```python
import numpy as np
import re
from collections import defaultdict
import networkx as nx
from tqdm import tqdm
from node2vec import Node2Vec
import pickle
import multiprocessing
from gensim.models import Word2Vec
import pandas
```

# Read graph

```python
G = nx.DiGraph()
```

### Load serialized graph

```python
with open('py_data/nx_fc_graph.pickle', 'rb') as f:
    G = pickle.load(f)
```

### Load csv and build new graph

```python
call_edges = pandas.read_csv('py_data/calls.csv')
```

```python
for i, edge in call_edges.iterrows():
    G.add_edge(edge.source_node_id, edge.target_node_id)
```

# Generate walks

```python
node2vec = Node2Vec(G, dimensions=128, walk_length=20, num_walks=30, workers=1)  # Use temp_folder for big graphs
```

```python
len(node2vec.walks)
```

```python
with open('py_data/walks.txt', 'w') as f:
    for walk in node2vec.walks:
        f.write(' '.join(walk) + '\n')
```

# Train word2vec

```python
import logging  # Setting up the loggings to monitor gensim
from time import time  # To time our operations
logging.basicConfig(format="%(levelname)s - %(asctime)s: %(message)s", datefmt= '%H:%M:%S', level=logging.INFO)
```

```python
model = Word2Vec(size = 128, window=5, min_count=0, workers = 4, sg = 1)
```

```python
model.build_vocab(node2vec.walks, progress_per=10000)
```

```python
t = time()

model.train(node2vec.walks, total_examples=model.corpus_count, epochs=30, report_delay=1)

print('Time to train the model: {} mins'.format(round((time() - t) / 60, 2)))
```

```python
in_vecs = []
out_vecs = []
for node_id, val in model.wv.vocab.items():
    id = val.index
    in_vecs.append((int(node_id), model.wv.vectors[id]))
    out_vecs.append((int(node_id), model.trainables.syn1neg[id]))
```

```python
in_vecs = pandas.DataFrame(in_vecs, columns=['id', 'in_vec']).set_index('id')
out_vecs = pandas.DataFrame(out_vecs, columns=['id', 'out_vec']).set_index('id')
```

```python
nodes = nodes.set_index('id')
```

```python
vecs = in_vecs.merge(out_vecs, left_index=True, right_index=True)
nodes = nodes.merge(vecs, left_index=True, right_index=True)
```
